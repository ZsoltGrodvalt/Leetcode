```python
import pandas as pd

def daily_leads_and_partners(daily_sales: pd.DataFrame) -> pd.DataFrame:
    lead = daily_sales[['date_id','make_name','lead_id']].drop_duplicates()
    grouped_lead= lead.groupby(['date_id','make_name']).count().reset_index()
    part = daily_sales[['date_id','make_name','partner_id']].drop_duplicates()
    grouped_part = part.groupby(['date_id','make_name']).count().reset_index()
    daily = pd.DataFrame({
        'date_id':grouped_lead['date_id'],
        'make_name':grouped_lead['make_name'],
        'unique_leads':grouped_lead['lead_id'],
        'unique_partners':grouped_part['partner_id'],
    })
    return daily
```
